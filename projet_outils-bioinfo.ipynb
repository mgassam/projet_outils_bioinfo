## Exercice 1 ([lien vers l'√©nonc√© original](https://python.sdv.univ-paris-diderot.fr/05_boucles_comparaisons/#5410-pyramide))

Ecrire un script qui dessine la pyramide repr√©sent√©e ci dessous :
```
         *
        ***
       *****
      *******
     *********
    ***********
   *************
  ***************
 *****************
*******************
```

````python
a = int(input("Veuilllez saisair le nombre de lignes :"))
for i in range(1,a+1):
    for j in range(1,a-i+1):
        print(" ",end="")
    for j in range(1,2*i-1+1):
        print("*",end="")
    print()
````
